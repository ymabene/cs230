```python

from sklearn.metrics import mean_absolute_error
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Comparing results for pretained model with scaling = 12,3,3,1

data = pd.read_csv('outputs\default_scaling\zinc_pretrained_inference.csv')
```

```python


pred_logp = data['pred_logp'].to_list()
true_logp = data['true_logp'].to_list()

pred_qed = data['pred_qed'].to_list()
true_qed = data['true_qed'].to_list()

pred_sas = data['pred_sass'].to_list()
true_sas = data['true_sass'].to_list()
```

```python

#MAE

mae = mean_absolute_error(true_logp, pred_logp)
print('Mean Absolute Error:', mae)
```

```python
#MAE

mae = mean_absolute_error(true_qed, pred_qed)
print('Mean Absolute Error:', mae)

```

```python

#MAE

mae = mean_absolute_error(true_sas, pred_sas)
print('Mean Absolute Error:', mae)

```

```python
## Plot distribution of properties

plt.hist(true_logp, bins = 30, alpha = 0.5, label ='True LogP Scores', color = 'blue' )
plt.hist(pred_logp, bins = 30, alpha = 0.5, label ='Predicted LogP Scores', color = 'orange' )
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of True and Predicted LogP Scores')
plt.legend()
plt.show()

```

```python

plt.hist(true_qed, bins = 30, alpha = 0.5, label ='True QED Scores', color = 'blue' )
plt.hist(pred_qed, bins = 30, alpha = 0.5, label ='Predicted QED Scores', color = 'orange' )

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of True and Predicted QED Scores')
plt.legend()
plt.show()
```

```python


plt.hist(true_sas, bins = 30, alpha = 0.5, label ='True SAS Scores', color = 'blue' )
plt.hist(pred_sas, bins = 30, alpha = 0.5, label ='Predicted SAS Scores', color = 'orange' )

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of True and Predicted SAS Scores')
plt.legend()
plt.show()
```

```python
data = pd.read_csv('outputs\zinc_pretrained_loss.csv') # Default scaling
```

```python

train_loss = data['train_loss']
val_loss = data['val_loss']

epochs = range(1,11)

plt.figure(figsize=(10,6))
plt.plot(epochs,train_loss, 'bo', label = 'Training Loss')
plt.plot(epochs,train_loss, 'ro', label = 'Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss (Proportional Property Scaling)')
plt.legend()
plt.show()

```

```python
## Inference on Molformer with biological data

data = pd.read_csv('outputs\\bio_1e-3\\bio_inference.csv')
```
