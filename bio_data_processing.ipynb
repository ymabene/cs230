```python
## Preprocessing of biological data for finetuning


import pandas as pd

```

```python
# Remove NA from chembl smiles

chembl_smiles = pd.read_csv('data\chembl_smiles_logp_qed.csv')

chembl_smiles = chembl_smiles.dropna()
```

```python
from rdkit import Chem
from rdkit.Chem import RDConfig
import os
import sys
sys.path.append(os.path.join(RDConfig.RDContribDir, 'SA_Score'))
import sascorer
```

```python

# Add qed scores

smiles = chembl_smiles['SMILES'].to_list()
sas_scores = []

for smile in smiles:
    mol = Chem.MolFromSmiles(smile)
    sas = sascorer.calculateScore(mol)
    sas_scores.append(sas)


```

```python
chembl_smiles['sas'] = sas_scores
```

```python
chembl_smiles.head()
```

```python
bio_data = pd.read_csv('data\processed_BRCA_data_with_pathways.csv')
```

```python
bio_data.head()
```

```python
## Merge target gene MAP3K1

mapk1_smiles = chembl_smiles[chembl_smiles['Gene'] == 'MAP2K1']
```

```python
mapk1_bio = bio_data[(bio_data['mutation_MAP3K1'] == 1) & (bio_data['mutation_PIK3CA'] == 0) &( bio_data['mutation_TP53'] == 0) ]
```

```python
import numpy as np
```

```python

mapk1_smiles_rep = pd.DataFrame(np.repeat(mapk1_smiles.values,len(mapk1_bio), axis = 0), columns= mapk1_smiles.columns)
```

```python
mapk1_bio_rep = pd.DataFrame(mapk1_bio.values.tolist() * len(mapk1_smiles), columns=mapk1_bio.columns)
```

```python
# Merge PIK3CA target gene

pik3_smiles = chembl_smiles[chembl_smiles['Gene'] == 'PIK3CA']
pik3_bio = bio_data[(bio_data['mutation_MAP3K1'] == 0) & (bio_data['mutation_PIK3CA'] == 1) &( bio_data['mutation_TP53'] == 0)]
pik3_smiles_rep = pd.DataFrame(np.repeat(pik3_smiles.values,len(pik3_bio), axis = 0), columns= pik3_smiles.columns)
pik3_bio_rep = pd.DataFrame(pik3_bio.values.tolist() * len(pik3_smiles), columns=pik3_bio.columns)
bio_chem_pik3 = pd.concat([pik3_smiles_rep, pik3_bio_rep], axis = 1)
```
